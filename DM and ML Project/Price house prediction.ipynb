```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
import matplotlib
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df = pd.read_csv("Downloads/Bengaluru_House_Data.csv")
df.head()
```

```python
df.shape
```

```python
df.groupby("area_type")["area_type"].unique().agg("count")
df.groupby
```

```python
df.isnull().sum()
```

```python
df2=df.drop(["society"], axis="columns")
df3=df2.dropna()
```

```python
df3.isnull().sum()
```

```python
df3.shape
```

```python
df3["size"].unique()
```

```python
df3["bhk"] = df3["size"].apply(lambda x : int(x.split(' ')[0]))
```

```python
df3.head()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
    
```

```python
df3[~df3["total_sqft"].apply(is_float)].head(50)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split("-")
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
df4 = df3.copy()
df4["total_sqft"] = df4["total_sqft"].apply(convert_sqft_to_num)
df4.head()
```

```python
df4.loc[30]
```

```python
df4.head()
```

```python
df4["new price"] = df4["price"]*100000
df4["price_per_sqft"] = df4["new price"]/df4["total_sqft"]
```

```python
df4.head()
```

```python
y = df4["area_type"].value_counts()
y
```

```python
df4.shape
```

```python
x = df4["location"].value_counts()
x
```

```python
df4.location = df4.location.apply(lambda x : x.strip())
location_stats = df4.groupby("location")["location"].agg("count").sort_values(ascending = False)
location_stats
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
```

```python
len(df4.location.unique())
```

```python
df4.location = df4.location.apply(lambda x : "other" if x in location_stats_less_than_10 else x)
len(df4.location.unique())
```

```python
df5 = df4[~(df4.total_sqft/df4.bhk<300)]
df5.shape
```

```python
df5.price_per_sqft.describe()
```

```python
def remove_outliers(df):
    df_out = pd.DataFrame()
    for key , x in df.groupby("location"):
        m = np.mean(x.price_per_sqft)
        st = np.std(x.price_per_sqft)
        reduced_df = x[(x.price_per_sqft>(m-st)) & (x.price_per_sqft <= (m+st))]
        df_out = pd.concat([reduced_df,df_out] ,ignore_index = True)
    return df_out

df6 = remove_outliers(df5)
df6.shape
df6.area_type.unique()
```

```python

dummies = pd.get_dummies(df6.location)
dummies.head
df7 = pd.concat([df6,dummies.drop("other",axis="columns")],axis="columns")


dummies2 = pd.get_dummies(df6.area_type)
df8 = pd.concat([df7,dummies2.drop("Carpet  Area",axis="columns")],axis="columns")
df9 = df8.drop(["area_type","availability","location","size"],axis="columns")
df9.head
```

```python
X = df9.drop(["price","new price","price_per_sqft"],axis="columns")
Y = df9.price
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test ,Y_train , Y_test = train_test_split(X,Y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train,Y_train)
clf.score(X_test,Y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(),X,Y, cv=cv)
```
